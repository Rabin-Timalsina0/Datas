```python
rating_books = ratings.merge(books,on='ISBN')
```

```python
rating_books
```

```python
num_rated_books = rating_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rated_books.rename(columns={'Book-Rating':'num-rating'},inplace=True)
num_rated_books
```

```python
avg_rated_books = rating_books.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rated_books.rename(columns={'Book-Rating':'avg-rating'},inplace=True)
avg_rated_books
```

```python
popular_df = num_rated_books.merge(avg_rated_books,on='Book-Title')
```

```python
popular_df
```

```python
popular_df = popular_df[popular_df['num-rating'] >= 250].sort_values('avg-rating',ascending=False).head(75)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','num-rating','avg-rating','Book-Author','Image-URL-M']]
```

```python
popular_df
```

```python
rating_books
```

```python
X = rating_books.groupby('User-ID')['Book-Rating'].count().reset_index().sort_values('Book-Rating',ascending=False)
X = X[X['Book-Rating']>200]
users_major = X['User-ID']
users_major
```

```python
filtered = rating_books[rating_books['User-ID'].isin(users_major)]
y = filtered.groupby('Book-Title')['Book-Rating'].count()>=50
famous_books = y[y].index
```

```python
  famous_books
```

```python
final_ratings = filtered[filtered['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend_book(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend_book('1984')
```
